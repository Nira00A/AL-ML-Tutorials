```python
class Insertion:
    def __init__(self,arr):
        self.arr = arr

    def __str__(self):
        return f'The given array is {self.arr}'

    def insertion_sort(self):
        arr = self.arr
        for i in range(1 , len(arr)):
            key = arr[i]
            j = i - 1
            while j>= 0 and arr[j] > key:
                arr[j+1] = arr[j]
                j -= 1

            arr[j+1] = key

        return arr
```

```python
i1 = Insertion([0,25,14,8,1,61,6])
i1.insertion_sort()
```

```python
class Bubble:
    def __init__(self , array):
        self.array = array

    def bubble_sort(self):
        n = len(self.array)
        while True:
            swap = False
            for i in range(n - 1):
                if self.array[i] > self.array[i+1]:
                    self.array[i] , self.array[i+1] = self.array[i+1] , self.array[i]
                    swap = True

            n -= 1

            if not swap:
                break            
```

```python
arr = [0,25,14,8,1,61,6]
b1 = Bubble(arr)
b1.bubble_sort()

print(arr)
```
