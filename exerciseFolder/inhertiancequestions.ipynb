```python
class Animal:
    def __init__(self , name , species):
        self.name = name
        self.species = species

class Dog(Animal):
    def __init__(self , name , species , breed):
        super().__init__(name , species)
        self.breed = breed

    def __str__(self):
        return str(self.name)

    def bark(self):
        return f'{self.name} says woof woof'
```

```python
d1 = Dog('bruno' , 'dog' , 'bulldog')
print(d1)
print(d1.bark())
```

```python
class Walker:
    def walk():
        return f'The person is walking'

class Runner:
    def run():
        return f'The person is running'

class Athlete(Walker , Runner):
    def __init__(self , training_hours):
        self.training_hours = training_hours

    def train(self):
        return f"The person is training for {self.training_hours}"
    
    def walk(self):
        return f"The athlete is walking"
```

```python
a1 = Athlete(10)
print(a1.walk())
print(a1.train())
```

```python
class A:
    def show(self):
        return f'Showing for A'

class B(A):
    def show(self):
        return f'Showing for B'

class C(A):
    def show(self):
        return f'Showing for C'

class D(B , C):
    pass

d = D()
print(d.show())
```

```python
class Vehicle:
    def start(self):
        return 'vehicle started'

class Car(Vehicle):
    def start(self):
        return super().start()
```

```python
v1 = Car()
print(v1.start())
```
