```python
import os

with open('example.txt' , 'w+') as file:
    file.write('Hey there\nIts a new message here')

    file.seek(0)
    
    content = file.read()
    print(content)
```

```python
with open('example.txt' , 'a+') as file:
    file.write('\nThis is an appended message')
    print('Message successfully appended')

    file.seek(0)
    
    content = file.read()
    print(content)
```

```python
with open('example.txt' , 'r') as file:
    content = file.read()
    new_content = content.replace("a" , "s")

with open('example.txt' , 'w') as file:
    file.write(new_content)
    print('Successfully updated')
    
```
