```python
'''Reading the whole file'''

with open('example.txt' , 'r') as file:
    content = file.read()
    print(content)
```

```python
'''Read a file line by line'''

with open('example.txt' , 'r') as file:
    for line in file:
        print(line.strip()) ##Removes the extra /new line character from the .txt file
```

```python
'''Writing a file'''

with open('example.txt' , 'w') as file:
    file.write('Hello world\n')
```

```python
'''Write a file without overwriting'''

with open('example.txt' , 'a') as file:
    file.write('This is an appended message\n')
```

```python
'''Writing many lines in a file with writelines'''
lines = ["\n1st line\n" , "2nd Line\n" , "3rd Line\n"]

with open('example.txt' , 'a') as file:
    file.writelines(lines)
```

```python
'''Working with binary file'''

data = b'\x00\x01\x02\x03\x04\x05'
with open('example.bin' , 'wb') as file:
    file.write(data)
```

```python
'''Reading a binary file'''
with open('example.bin' , 'rb') as file:
    print(file.read())
```

```python
'''Read content from a .txt file to another .txt file'''
with open('example.txt' , 'r') as file:
    content = file.read()

with open('destination.txt' , 'w') as destination_file:
    destination_file.write(content)

'''
with statement
The with open(...) as file: block is used for context management — it opens a file and ensures it’s automatically closed when the block ends.
But it doesn’t create a new scope in Python.

Variable scope in Python
In Python, only functions, classes, and modules create their own scope — not with, for, or if statements.
That means any variable you assign inside a with block still belongs to the current (outer) scope — in this case, the global scope (if your code is at the top level of the script).

So content works here because:

You defined it inside the with block.

The with block ended, but the variable still exists in the same scope.

You can still use it afterwards in the second with block.
'''
```

```python
'''Reading a text file and count the numbers of lines , words and characters'''

with open('example.txt' , 'r') as file:
    content = file.read()

    lines = content.splitlines()
    print(len(lines))

    words = content.split()
    print(len(words))

    total_char = sum(map(len, words))
    print(total_char)
```

```python
'''Writing and reading'''

with open('example.txt' , 'w+') as file:
    file.write("Hello world\n")
    file.write("This is a new line \n")

    ## Move the file cursor to the beginning
    file.seek(0)

    ## Read the content of the file
    content = file.read()
    print(content)
```

```python
'''Checking if the path is a file , directory or anything else'''
import os

path = 'example.txt'

if os.path.isfile(path):
    print(f"{path} is a file.")
elif os.path.isdir(path):
    print(f"{path} is a directory.")
else:
    print(f"{path} is something else.")
```

```python
'''Getting absolute path'''
relative_path = "example.txt"
absolute_path = os.path.abspath(relative_path)
print(f"the absolute path of {relative_path} is \n{absolute_path}")
```
