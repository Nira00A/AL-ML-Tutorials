```python
connection = sqlite3.connect('test.db')
```

```python
## Making a cursor
cursor = connection.cursor()
```

```python
cursor.execute('''CREATE TABLE employee (
id INT PRIMARY KEY,
name TEXT NOT NULL,
age INT NOT NULL,
department TEXT
)''')
```

```python
cursor.execute('''INSERT INTO employee (id , name , age , department) VALUES (1 ,'ArinOrin',20,'Bsc')''')
connection.commit()
```
