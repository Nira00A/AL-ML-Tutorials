```python
'''Question1'''

class Shape:
    def area():
        return 'Area of the shape'

class Circle(Shape):
    def __init__(self , radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

c1 = Circle(10)
print('The area is' , c1.area())
```

```python
'''Question2'''
def describe_shape(shape):
    return shape.area()

describe_shape(Circle(10))
```

```python
'''Question3'''
from abc import ABC , abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        return f'Car engine started'

v1 = Car()
print(v1.start_engine())
```

```python
'''Question7'''
class Error(Exception):
    pass

class Person(Error):
    def __init__(self , name , age):

        try:
            if age > 30:
                print('Age verified')
            else:
                raise Error
        except Error:
            return 'Age Exception occured'
        
        self.name = name
        self.age = age
        
    def get(self):
        return f'Name : {self.name}\nAge : {self.age}'

    def set(self , name , age):
        self.name = name
        self.age = age
        return f"Updated Name : {self.name}\nUpdated Age : {self.age}"

class Student(Person):
    def __init__(self , name , age , stdId):
        super().__init__(name , age)
        self.stdId = stdId

    def __str__(self):
        try:
            return f'Student ifo is : \nName:{self.name}\nAge:{self.age}'
        except AttributeError:
            return 'Cannot create the Student class'
```
