```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

#Function to plot PMF AND CDF
def plot_discrete_dist(dist_name , x_values , pmf_values , cdf_values):
    plt.figure(figsize=(14,8))
    
    #plot description
    plt.subplot(1,2,1)
    plt.stem(x_values , pmf_values)
    plt.title('PMF')
    plt.grid(True)

    plt.subplot(1,2,2)
    plt.step(x_values , cdf_values)
    plt.title('CDF')
    plt.grid(True)
```

```python
dist_name = 'Binomial'
n = int(input("Enter the number of n trials :"))
p = 0.5

x_values = np.arange(n+1)
pmf_values = binom.pmf(x_values,n,p)
cdf_values = binom.cdf(x_values,n,p)

plot_discrete_dist(dist_name , x_values , pmf_values , cdf_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_dist(k_values , pmf_values , cdf_values):
    plt.figure(figsize= (10,4))

    #pmf
    plt.subplot(1,2,1)
    plt.stem(k_values , pmf_values)
    plt.title(label="PMF OF POISSON")

    #cdf
    plt.subplot(1,2,2)
    plt.stem(k_values , cdf_values)
    plt.title(label="CDF OF POISSON")


k_values = np.arange(0 , 10)
lamda = 0.3
pmf_values = poisson.pmf(k=k_values , mu=lamda)
cdf_values = poisson.cdf(k=k_values , mu=lamda)

plot_poisson_dist(k_values , pmf_values=pmf_values , cdf_values=cdf_values)
```

```python
from scipy.stats import norm

def plot_continuous_dist(dist_name , x_values , pdf_values , cdf_values):
    plt.figure(figsize=(14,8))
    
    #plot description
    plt.subplot(1,2,1)
    plt.plot(x_values , pdf_values)
    plt.title('PDF')
    plt.grid(True)

    plt.subplot(1,2,2)
    plt.plot(x_values , cdf_values)
    plt.title('CDF')
    plt.grid(True)

mu = 0
sigma = 1
x_values = np.linspace(mu-4*sigma , mu+4*sigma)
norm_pdf = norm.pdf(x_values , mu , sigma)
norm_cdf = norm.cdf(x_values , mu , sigma)

plot_continuous_dist("Normal" , x_values , norm_pdf , norm_cdf)
```

```python
#Uniform 
from scipy.stats import uniform

#Scipy Method
a = 0
b = 15
x_values = np.linspace(0 , 15)
uni_pdf = uniform.pdf(x_values , a , b)
uni_cdf = uniform.cdf(x_values , a , b)

plot_continuous_dist("Uniform" , x_values , pdf_values=uni_pdf , cdf_values=uni_cdf)
```
