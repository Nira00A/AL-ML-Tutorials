```python
tpl = 1,2,3,4,1,2,4,6,1,4,2,1,2

count = {}
for item in tpl:
    if item in count:
        count[item] += 1
    else:
        count[item] = 1
print(count)
```

```python
def print_nested_tuple_items(tpl):
    for item in tpl:
        if type(item) is tuple:
            print_nested_tuple_items(item)
        else:
            print(item)

nested_tpl = (1, (2, 3), (4, (5, 6, (7, (8,(9,0,(12,57)))))), 9)
print_nested_tuple_items(nested_tpl)
```
