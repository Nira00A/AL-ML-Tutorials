```python
dict1 = {}

def str_dict(str):
    str_lst = list(str.lower())
    for item in str_lst:
        if item in dict1:
            dict1[item] += 1
        else:
            dict1[item] = 1
    return dict1

print(str_dict("OrinHorin"))
```

```python
def rotate_list(lst, k):
    for item in range(k):
        back_items = lst.pop()
        lst.insert(0,back_items)
    return lst

rotate_list([10 , 20 , 30 , 40 , 50] , 3)
```
