```python
def fibonacci_series(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_series(n - 1) + fibonacci_series(n - 2)

const = fibonacci_series(3)
print(const)
```

```python
def nested_default_argumnts(b = None , *a):
    if b is None:
        b = {}
    for x,y in a:
        b[x] = y
    return b

const = nested_default_argumnts( None ,('key' ,'value'))
print(const)
```

```python
def variable_keyward_arguments(*pairs):
    b = {}
    for x , y in pairs:
        b[x] = y
    return b

variable_keyward_arguments(('key1' , 1),('key2' , 2))
```

```python
def calculate_lift_rounds(n, capacity):
    """
    Function to calculate the number of rounds the lift needs to cover.
    
    Parameters:
    n (int): Total number of people.
    capacity (int): Maximum number of people the lift can carry in one round.
    
    Returns:
    int: The number of rounds required to transport all people to the top floor.
    """
    # Your code here
    if n % capacity == 0:
        return n // capacity
    else:
        return (n // capacity) + 1
   
    

calculate_lift_rounds(7 , 4)
```

```python
lis = [1,2,3,4,5,6]

def check_even_odd(lis):
    even = 0
    odd = 0

    for x in lis:
        if x % 2 ==0:
            even += 1
        else:
            odd += 1
    return even , odd
```

```python
lst1 = [1, 2 , 3]
lst2 = [1, 2, 3, 4, 5]

def is_subset(lst1, lst2):
    set1 = set(lst1)
    set2 = set(lst2)
    
    if set1.issubset(set2): 
        return True
    else: 
        return False

is_subset(lst1 , lst2)
```

```python
def rotate_list(lst, k):
    l = []
    for i in range(k  , len(lst)):
        l.append(lst[i])
        
    for i in range(k -1):
        l.append(lst[i])
        
    return l

rotate_list([10, 20, 30, 40, 50] , 3)
```
