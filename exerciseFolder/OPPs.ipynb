```python
'''Operation Overloading'''

class Complex:
    def __init__(self , real , comp):
        self.real = real
        self.comp = comp

    def __str__(self):
        return f"{self.real} + {self.comp}i"

    def __add__(self , other):
        return complex(self.real + other.real , self.comp + other.comp)

    def __sub__(self , other):
        return complex(self.real - other.real , self.comp - other.comp)
```

```python
c1 = Complex(10 , 10)
c2 = Complex(1 , 1)

print(c1)
print(c2)
```

```python
s1 = c1.__add__(c2)
s2 = c2.__sub__(c1)
print(s1 , s2)
```
