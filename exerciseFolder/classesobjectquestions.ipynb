```python
class Employee:
    def __init__(self , name , salary):
        self.name = name
        self.salary = salary

    def __str__(self):
        return f"Current salary of {self.name} is {self.salary}"

class Manager(Employee):
    def __init__(self ,name , salary):
        Employee.__init__(self , name , salary)

    def __str__(self):
        return f"Current salary of {self.name} is {self.salary}"
    
    def bonus(self , bonus):
        self.salary += bonus
        return self.salary
```

```python
E1 = Manager('Bacchus' , 100000)
E2 = Employee('OrinHorin' , 20000)

print(f'{E1}\n{E2}')
```

```python
E1.bonus(20000)
```

```python
'''Abstruction and Polymorisation'''

from abc import ABC , abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print('bark')

class Cat(Animal):
    def sound(self):
        print('Meow')

def animal_sounds(animal_list):
    for animal in animal_list:
        animal.sound()

animals = [Dog() , Cat() , Dog()]
animal_sounds(animals)
```

```python
from abc import ABC , abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        return "start engine"

    def stop_engine(self):
        return "stop engine"

class Car(Vehicle):
    def start_engine(self):
        return "Engine started for car"

    def stop_engine(self):
        return "Engine stoped for car"

class Boat(Vehicle):
    def start_engine(self):
        return "Engine started for boat"

    def stop_engine(self):
        return "Engine stoped for boat"

class Plane(Vehicle):
    def start_engine(self):
        return "Engine started for plane"

    def stop_engine(self):
        return "Engine stoped for plane"

def polimo(vehicle_list):
    list(map(lambda x:print(f'{x.start_engine()} and {x.stop_engine()}') , vehicle_list))

vehicle_list = [Car() , Boat() , Plane()]
print(polimo(vehicle_list))
```

```python
from abc import ABC , abstractmethod

class Payment(ABC): 
    def __init__(self , amount):
        self.amount = amount
    
    @abstractmethod
    def payment(self , amount):
        pass

class Paypal(Payment):
    def __init__(self , amount):
        Payment.__init__(self , amount)
    def payment(self , amount):
        amount += amount
        return amount

class Bank(Payment):
    def __init__(self , amount):
        Payment.__init__(self , amount)
    def payment(self , amount):
        amount += amount
        return amount

class Bitcoin(Payment):
    def __init__(self , amount):
        Payment.__init__(self , amount)
    def payment(self , amount):
        amount += amount
        return amount

p1 = Paypal(1000)
print(p1.payment(1000))

p2 = Bank(1500)
print(p2.payment(1500))
```
