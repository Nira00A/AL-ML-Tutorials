```python
## Building a QuadTree
class Point:
    def __init__(self , x , y):
        self.x = x
        self.y = y

class Node: 
    def __init__(self , pos , data):
        self.pos = pos
        self.data = data

class Quad:
    def __init__(self , topL , botR):
        self.topL = topL
        self.botR = botR
        self.n = None
        self.topLeft = None
        self.topRight = None
        self.botLeft = None
        self.botRight = None

    def insert(self , node):
        if node is None:
            return None
        
        if not self.isBounding(node):
            return None
        
        if self.n is None and self.topLeft is None:
            self.n = node
            return True

        mid_x = (self.topL.x + self.botR.x) / 2
        mid_y = (self.topL.y + self.botR.y) / 2

        if node.x <= mid_x:
            if node.y <= mid_y:
                if self.topLeft is None:
                    self.topLeft = Quad(self.topL, Point(mid_x, mid_y))
                
                self.topLeft.insert(node)
                return self.topLeft

                
    def search(self , node):
        if node is None:
            return None
        
        if not self.isBounding(node):
            return None
        
        if self.n is not None:
            return self.n
        
        if (self.topL.x + self.botR.x) / 2 >= node.x :
            if (self.topL.y + self.botR.y) / 2 >= node.y:
                if self.topLeft is None:
                    return None
                return self.topLeft.search(node)
            else:
                if self.botLeft is None:
                    return None 
                return self.botLeft.search(node)
        else:
            if (self.topL.y + self.botR.y) / 2 < node.y:
                if self.topRight is None:
                    return None
                return self.topRight.search(node)
            else:
                if self.botRight is None:
                    return None
                return self.botRight.search(node)
        
            

    def isBounding(self , pos):
        return (self.topL.x <= pos.x and
                self.topL.y <= pos.y and 
                self.botR.x > pos.x and 
                self.botR.y > pos.y)
```
