```python
# Bubble sort
# Time: O(n^2)
# Space: O(1)

A = [-4 ,4 ,-9, 23, 4, 3, -7 , 0 , 1 , 5]

def bubble_sort(arr):
    n = len(arr)
    flag = True
    while flag:
        flag = False
        for i in range(1,n):
            if arr[i-1] > arr[i]:
                flag = True
                arr[i-1] , arr[i] = arr[i] , arr[i-1]

    return arr

print(bubble_sort(A))
```

```python
# Insertion sort
# Time: O(n^2)
# Space: O(1)

B = [-4 ,-10 ,-9, 23, 4, 3, -7 , 0 , 1 , 5]

def insertion_sort(arr):
    n = len(arr)
    for i in range(1 , n):
        for j in range(i , 0 , -1):
            if arr[j-1] > arr[j]:
                arr[j-1] , arr[j] =  arr[j] , arr[j-1]
            else:
                break
    return arr

print(insertion_sort(B))
```

```python
# Selection sort
# Time: O(n^2)
# Space: O(1)

C = [-4 ,-10 ,-9, 23, 4, 3, -7 , 0 , 1 , 5]

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1 , n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i] , arr[min_idx] = arr[min_idx], arr[i]
    
    return arr

print(selection_sort(C))
```
