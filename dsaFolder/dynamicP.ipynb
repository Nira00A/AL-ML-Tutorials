```python
arr = [8,2,3,4]

def solve_mncs(arr):
    n = len(arr)
    
    if n == 0: return

    if n == 1: return 

    dp = [0] * n

    dp[0] = arr[0]
    dp[1] = max(dp[0] , arr[1])

    for i in range(2 , n):
        dp[i] = max(dp[i-1] , arr[i] + dp[i-2])

    return dp[n-1]

print(solve_mncs(arr=arr))
```

```python
## formula for Knapsack Problem (0/1) based

## m[i , w] = max(m[i-1,w] , m[i-1 , w-w[i]] + p[i]) // p is profit and i each item and weight is weight give for each item
 
import numpy as np

def knapsack(weights , values , capacity):
    n = len(values)

    dp = [[0 for _ in range(capacity + 1)] for _ in range(n+1)]

    for i in range(1 , n+ 1):
        for w in range(1 , capacity + 1):
            curr_val = values[i-1]
            curr_wet = weights[i-1]

            if curr_wet <= w:
                dp[i][w] = max(dp[i-1][w] , curr_val + dp[i-1][w - curr_wet])
            else:
                dp[i][w] = dp[i-1][w]

    return dp[n][capacity]

values = [3, 2, 4, 4]
weights = [4, 3, 2, 3]
capacity = 6

print(knapsack(weights , values , capacity))
```

```python
## for Scheduling Jobs

## done by calculating the Ratio (w/l) for each.

def scheduling(jobs):
    sorted_optimum_solution = sorted(jobs, key=lambda x:(x['w'] / x['l'])  ,reverse=True)

    total_time = 0
    curr_time = 0

    for job in sorted_optimum_solution:
        curr_wet = job['w']
        curr_len = job['l']

        curr_time += curr_len
        cost = curr_wet * curr_time 

        total_time += cost

    return sorted_optimum_solution , total_time

jobs_data = [
    {'id': 'J1', 'l': 6, 'w': 12},
    {'id': 'J2', 'l': 2, 'w': 8},
    {'id': 'J3', 'l': 3, 'w': 9}
]

final_cost = scheduling(jobs_data)
print(final_cost)
```

```python
def solve_scheduling(jobs):
    # jobs is a list of dictionaries: {'id': 'J1', 'l': 6, 'w': 12}
    
    # 1. Calculate Ratios and Sort
    # We want HIGHEST ratio first, so we use reverse=True
    # key=lambda x: x['w'] / x['l'] tells Python to sort based on that division
    optimal_schedule = sorted(jobs, key=lambda x: x['w'] / x['l'], reverse=True)
    
    # 2. Calculate Total Cost
    current_time = 0
    total_weighted_cost = 0
    
    print("Optimal Order:")
    for job in optimal_schedule:
        current_time += job['l']  # Add job length to current clock
        cost = job['w'] * current_time
        total_weighted_cost += cost
        
        print(f"Job {job['id']}: Ratio {job['w']/job['l']:.2f}, "
              f"Finishes at t={current_time}, Cost={cost}")
        
    return total_weighted_cost

# --- Test with the Practice Sum Data ---
jobs_data = [
    {'id': 'J1', 'l': 6, 'w': 12},
    {'id': 'J2', 'l': 2, 'w': 8},
    {'id': 'J3', 'l': 3, 'w': 9}
]

final_cost = solve_scheduling(jobs_data)
print(f"\nTotal Weighted Completion Time: {final_cost}")
```
