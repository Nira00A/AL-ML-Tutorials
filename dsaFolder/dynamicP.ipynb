```python
arr = [8,2,3,4]

def solve_mncs(arr):
    n = len(arr)
    
    if n == 0: return

    if n == 1: return 

    dp = [0] * n

    dp[0] = arr[0]
    dp[1] = max(dp[0] , arr[1])

    for i in range(2 , n):
        dp[i] = max(dp[i-1] , arr[i] + dp[i-2])

    return dp[n-1]

print(solve_mncs(arr=arr))
```
