```python
import numpy as np
import matplotlib as mlt
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.cluster import KMeans

def floyd_steinberg_dither(image, palette=None):
    """
    Apply FloydSteinberg error diffusion dithering.
    """
    # 2.1 Convert pixels to float for precise error calculations
    arr = image.astype(np.float32)
    height, width = arr.shape[:2]
    
    # 2.2 Detect if image is grayscale (2D) or color (3D)
    if len(arr.shape) == 2:
        is_color = False
        channels = 1
    else:
        is_color = True
        channels = arr.shape[2]

    # 2.3 Process each pixel in scanline order
    for y in range(height):
        for x in range(width):
            old = arr[y, x].copy()                # Original (with accumulated error)
            new = find_nearest_color(old, palette)
            arr[y, x] = new                       # Replace pixel with quantized value
            error = old - new                     # Compute quantization error
            
            # 2.4 Distribute error to neighbors
            if x+1 < width:
                arr[y, x+1] += error * 7/16
            if y+1 < height:
                if x>0:
                    arr[y+1, x-1] += error * 3/16
                arr[y+1, x]   += error * 5/16
                if x+1 < width:
                    arr[y+1, x+1] += error * 1/16
    
            # 2.5 Clip pixel values back to valid 0255 range and convert to uint8
            arr = np.clip(arr, 0, 255)
            return arr.astype(np.uint8)

def find_nearest_color(pixel, palette):
    """Return closest color from the palette (or black/white if no palette)."""
    if palette is None:
        if is_color:
            gray_val = np.mean(pixel)
            return np.array([0,0,0]) if gray_val < 128 else np.array([255,255,255])
        else:
            return 0.0 if pixel < 128 else 255.0
    
    # For color images, compute Euclidean distance to each palette color
    distances = np.sum((palette - pixel)**2, axis=1)
    return palette[np.argmin(distances)]

def kmeans_with_dithering(image, n_colors=8):
    height, width, channels = image.shape
    pixels = image.reshape(-1, channels)       # Flatten to list of pixels

    # 3.1 Fit KMeans to find 'n_colors' cluster centers
    kmeans = KMeans(n_clusters=n_colors, random_state=42)
    kmeans.fit(pixels)

    # 3.2 Quantize: replace each pixel by its cluster center
    quantized_pixels = kmeans.cluster_centers_[kmeans.labels_]
    quantized_image  = quantized_pixels.reshape(height, width, channels)

    # 3.3 Dither the quantized image to smooth bands
    palette = kmeans.cluster_centers_.astype(np.uint8)
    dithered = floyd_steinberg_dither(quantized_image, palette)

    return quantized_image.astype(np.uint8), dithered, palette
```

```python
import matplotlib.image as mpimg

image = mpimg.imread("Picture-Files/dog.jpg")

quantized, dithered, palette = kmeans_with_dithering(image, n_colors=2)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.imshow(quantized)
plt.axis('off')
plt.title('Quantized')
plt.subplot(1,2,2)
plt.imshow(dithered)
plt.axis('off')
plt.title('Dithered')
plt.show()
```

```python
import numpy as np
from PIL import Image

GREYSCALE = False
img_name = 'Picture-Files/dog.jpg'

# Read in the image, convert to greyscale.
img = Image.open(img_name)
if GREYSCALE:
    img = img.convert('L')

width, height = img.size
new_width = 400
new_height = int(height * new_width / width)
img = img.resize((new_width, new_height), Image.Resampling.LANCZOS)

def get_new_val(old_val, nc):
    """
    Get the "closest" colour to old_val in the range [0,1] per channel divided
    into nc values.

    """

    return np.round(old_val * (nc - 1)) / (nc - 1)

# For RGB images, the following might give better colour-matching.
#p = np.linspace(0, 1, nc)
#p = np.array(list(product(p,p,p)))
#def get_new_val(old_val):
#    idx = np.argmin(np.sum((old_val[None,:] - p)**2, axis=1))
#    return p[idx]

def fs_dither(img, nc):
    """
    Floyd-Steinberg dither the image img into a palette with nc colours per
    channel.

    """

    arr = np.array(img, dtype=float) / 255

    for ir in range(new_height):
        for ic in range(new_width):
            # NB need to copy here for RGB arrays otherwise err will be (0,0,0)!
            old_val = arr[ir, ic].copy()
            new_val = get_new_val(old_val, nc)
            arr[ir, ic] = new_val
            err = old_val - new_val
            # In this simple example, we will just ignore the border pixels.
            if ic < new_width - 1:
                arr[ir, ic+1] += err * 7/16
            if ir < new_height - 1:
                if ic > 0:
                    arr[ir+1, ic-1] += err * 3/16
                arr[ir+1, ic] += err * 5/16
                if ic < new_width - 1:
                    arr[ir+1, ic+1] += err / 16

    carr = np.array(arr/np.max(arr, axis=(0,1)) * 255, dtype=np.uint8)
    return Image.fromarray(carr)


def palette_reduce(img, nc):
    """Simple palette reduction without dithering."""
    arr = np.array(img, dtype=float) / 255
    arr = get_new_val(arr, nc)

    carr = np.array(arr/np.max(arr) * 255, dtype=np.uint8)
    return Image.fromarray(carr)

for nc in (2, 3, 4, 8, 16):
    print('nc =', nc)
    dim = fs_dither(img, nc)
    dim.save('dimg-{}.jpg'.format(nc))
    rim = palette_reduce(img, nc)
    rim.save('rimg-{}.jpg'.format(nc))
```

```python
from sklearn.cluster import KMeans
from PIL import Image
import numpy as np
import numba

@numba.njit
def find_nearest_color(pixel, palette):
    min_dist = 1e10
    index = 0
    for i in range(palette.shape[0]):
        dist = 0.0
        for c in range(pixel.shape[0]):
            diff = pixel[c] - palette[i, c]
            dist += diff * diff
        if dist < min_dist:
            min_dist = dist
            index = i
    return palette[index]

@numba.njit
def floyd_steinberg_kmeans(image, palette):
    Lx, Ly, Lc = image.shape
    for j in range(Ly):
        for i in range(Lx):
            old_pixel = image[i,j,:].copy()
            new_pixel = find_nearest_color(old_pixel, palette)
            error = old_pixel - new_pixel
            image[i,j,:] = new_pixel
            if i < Lx-1:
                image[i+1,j,:] += error * 7/16
            if j < Ly-1:
                image[i,j+1,:] += error * 5/16
                if i > 0:
                    image[i-1,j+1,:] += error * 1/16
                if i < Lx-1:
                    image[i+1,j+1,:] += error * 3/16
    return image

# Usage:
image = Image.open('Picture-Files/dog1.jpg')
image_np = np.array(image)
flat_image = image_np.reshape(-1, 3)
kmeans = KMeans(n_clusters=4).fit(flat_image)
palette = kmeans.cluster_centers_.astype(np.float32)
dithered_img = floyd_steinberg_kmeans(image_np.astype(np.float32), palette)

dithered_img_clipped = np.clip(dithered_img, 0, 255).astype(np.uint8)
plt.imshow(dithered_img_clipped)
plt.show()
```
