```python
#Knn Manual in 28/11/2025
import numpy as np
from collections import Counter

class KNN:
    def __init__(self , k=None):
        self.k = k

    def getNeighbour(self,X,y,X_test):
        def eucledeon_distance(point1 , point2):
            return np.sqrt(np.sum((point1 - point2)**2))
        
        distance = [(eucledeon_distance(X[i],X_test) , y[i]) for i in range(len(X))]
        distance.sort(key=lambda x:x[0])
        
        return [label for _ , label in distance[:self.k]]

    def predict(self,X,y,X_test):
        ngb = self.getNeighbour(X,y,X_test)
        return Counter(ngb).most_common(1)[0][0]
```
