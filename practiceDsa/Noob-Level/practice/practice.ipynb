```python
#Question1 : You are given an array people where people[i] is the weight of 
# the ith person, and an infinite number of boats where each boat can carry 
# a maximum weight of limit. Each boat carries at most two people at the same 
# time, provided the sum of the weight of those people is at most limit.

# Return the minimum number of boats to carry every given person.

# Example 1:

# Input: people = [1,2], limit = 3
# Output: 1
# Explanation: 1 boat (1, 2)

def boat(arr,n):
    stk = []
    total_boat = 0

    for i in arr:
        if not stk:
            stk.append(i)
        
        elif stk[0] == n:
            total_boat += 1
            stk.pop()
        
        elif stk[0] + i > 3:
            stk.pop()
            total_boat += 1
            stk.append(i)
        
        else:
            stk[0] += i
    
    return total_boat

print(boat([3,2,2,1],3))
        
## Wrong approch
```

```python
def boat(people: list , limit: int):
    people.sort()

    l=0
    r=len(people) - 1
    total_boats = 0

    while l <= r:
        if people[l] + people[r] <= limit:
            l += 1
            r -= 1
        else:
            r-= 1

        total_boats += 1
    
    return total_boats

print(boat([3,5,5,10],3))
```

```python
#Question2 : Given two strings s and p, return an array of all the start indices of 
# p's anagrams in s. You may return the answer in any order.
# Example 1:

# Input: s = "cbaebabacd", p = "abc"
# Output: [0,6]
# Explanation:
# The substring with start index = 0 is "cba", which is an anagram of "abc".
# The substring with start index = 6 is "bac", which is an anagram of "abc".

def findAnagrams(s: str, p: str):
    # Edge case: If the recipe is longer than the string, it's impossible.
    if len(p) > len(s):
        return []

    p_count = {}
    window_count = {}
    res = []

    for i in range(len(p)):
        p_count[p[i]] = p_count.get(p[i] , 0) + 1
        window_count[s[i]] = window_count.get(s[i] , 0) + 1

    if p_count == window_count:
        res.append(0)
    
    l = 0
    for r in range(len(p) , len(s)):
        window_count[s[r]] = window_count.get(s[r] , 0) + 1

        window_count[s[l]] -= 1

        if (window_count[s[r]] == 0):
            del window_count[s[l]]

        l += 1

        if window_count == p_count:
            res.append(l)

    return res
    
print(findAnagrams("abcabc", "abc"))
# Output: [0, 6]
```
