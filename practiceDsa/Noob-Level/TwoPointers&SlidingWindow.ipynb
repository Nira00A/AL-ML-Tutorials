```python
#Question1: The Task: Given an array [1, 2, 3, 4, 5], 
# reverse it in place so it becomes [5, 4, 3, 2, 1].

def reverse(arr):
    l = 0
    r = len(arr) - 1

    while l < r:
        if arr[l] <= arr[r]:
            arr[l] , arr[r] = arr[r] , arr[l]
            l += 1
            r -= 1
    
    return arr

print(reverse([1,2,3,4,5]))
```

```python
#Question2: The Task: You have a sorted array [1, 3, 4, 7, 10] 
# and a target sum of 11.
#  Find the two numbers that add up to 11.

def smallSum(arr , target):
    l = 0
    r = len(arr) - 1

    while l < r:
        if (arr[l] + arr[r]) == target:
            return 'GotYa'
            
        elif (arr[l] + arr[r]) < target:
            l += 1
        else:
            r -= 1

    return "NahNah"

print(smallSum([1 , 2, 3 ,5 , 10] , 11))

#Note : Works only when sorted
```

```python
#Question2 : Works even when unsorted

def smallSumUnsorted(arr , target):
    seen = {}

    for i , n in enumerate(arr):
        compliment = target - n

        if compliment in seen:
            return [compliment , n]
        
        seen[n] = i
    
    return []

print(smallSumUnsorted([1, 10, 2, 3, 23, 5], 11))
```

```python
#Question3 : 
# Problem: The "Bouncer" (Move Zeros)
# Input: [0, 1, 0, 3, 12]
# Goal: Move zeros to the end in-place (no extra dictionary, no extra list).
# Constraint: You can't use sort() because the order of the numbers 
# 1, 3, 12 must stay the same.

def bouncer(arr):
    fast = 0
    slow = 0
    n = len(arr) - 1

    while fast <= n:
        if arr[fast] == 0:
            fast += 1
        else:
            arr[slow] , arr[fast] = arr[fast] , arr[slow]
            slow += 1
    
    return arr

print(bouncer([0, 1 ,5, 23,8 ,46 ,536 ,0, 6,1 ,0, 0, 3, 12]))
```

```python
#Question4: Dublicates

def dublicates(arr: list):
    sset = set()

    for i in arr:
        if i in sset:
            arr.remove(i)
        else:
            sset.add(i)
    
    return arr

print(dublicates([1,1,1,3,3,5]))
```

```python
def remove_duplicates_inplace(arr):
    if not arr: return 0
    
    # The 'slow' pointer: marks the position of the last unique element found
    slow = 0
    
    # The 'fast' pointer: scouts the rest of the array
    for fast in range(1, len(arr)):
        # If the scout finds a NEW unique value
        if arr[fast] != arr[slow]:
            # Move the slow pointer forward to the next available slot
            slow += 1
            # Copy the new unique value to the slow pointer's position
            arr[slow] = arr[fast]
            
    # The unique elements are now at the front of the array (up to index slow)
    return arr[:slow + 1]

# Note: This requires the array to be SORTED first!
print(remove_duplicates_inplace([1,1,1,3,3,5]))
# Output: [1, 2, 3, 5]
```

```python
#Sliding window Problems

#Question5: Longest Subsstring Without Repeating Characters
# Example 1:
# Input: s = "abcabcbb"
# Output: 3
# Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.

# Example 2:
# Input: s = "bbbbb"
# Output: 1
# Explanation: The answer is "b", with the length of 1.

def LongestSubsstringWithoutRepeating(str : str):
    longest = 0
    l = 0
    n = len(str)
    sset = set()

    for r in range(n):
        while str[r] in sset:
            sset.remove(str[l])
            l += 1

        w = (r - l) + 1
        longest = max(longest , w)
        sset.add(str[r])

    return longest

print(LongestSubsstringWithoutRepeating("acbcag"))
```

```python
#Question6 : You have two sorted arrays, A = [1, 3, 5] and B = [2, 4, 6].
#  Merge them into a single sorted array [1, 2, 3, 4, 5, 6] in $O(N)$ time.

def merge_two_sorted_ar(arr1 , arr2):
    p1 = 0
    p2 = 0
    new = []

    while p1 < len(arr1) and p2 < len(arr2):
        if arr1[p1] < arr2[p2]:
            new.append(arr1[p1])
            p1 += 1

        else:
            new.append(arr2[p2])
            p2 += 1

    if p1 < len(arr1): 
        new.extend(arr1[p1:])
    else: 
        new.extend(arr2[p2:])

    return new

print(merge_two_sorted_ar([1,3,5] , [2,4,6]))
```

```python
#Question7 :  Given an array [1, 12, -5, -6, 50, 3] and k = 4, 
# find the contiguous subarray of length 4 that has the maximum average value.

def sum_of_contiguous(arr: list):
    k = 4
    slow = 0
    total_sum = 0

    max_sum = float('-inf')

    if len(arr) < k:
        return sum(arr)

    for i in range(len(arr)):
        w = (i - slow) + 1
        if w < k:
            total_sum += arr[i]
        else:
            total_sum += arr[i]
            max_sum = max(max_sum , total_sum)
            total_sum -= arr[slow]
            slow += 1

    return max_sum

print(sum_of_contiguous([1 , 12 , -5 , -6 , 50 , 3]))
```

```python
#Question8 : : You are given an array of heights [1,8,6,2,5,4,8,3,7]. 
# Imagine these are vertical lines. Find two lines that, together with the x-axis, 
# form a container that holds the most water.

def maxArea(arr):
    l = 0
    r = len(arr) - 1
    maxA = 0

    while l < r:
        area = 0
        if arr[l] <= arr[r]:
            area = arr[l] * (r - l)
            l += 1
        else:
            area = arr[r] * (r - l)
            r -= 1
        
        maxA = max(maxA , area)

    return maxA

print(maxArea([1,8,6,2,5,4,8,3,7]))
```
