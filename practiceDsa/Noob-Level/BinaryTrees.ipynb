```python
class TreeNode:
    def __init__(self , node , left=None , right=None) -> None:
        self.node = node
        self.left = left
        self.right = right

    def __repr__(self) -> str:
        return str(self.node)
    

#         2
#      12   15
#     3   2    10
#   50 3 1 2  5  5

A = TreeNode(2)
B = TreeNode(12)
C = TreeNode(3)
D = TreeNode(50)
E = TreeNode(3)
F = TreeNode(2)
G = TreeNode(15)
H = TreeNode(10)
I = TreeNode(1)
J = TreeNode(2)
K = TreeNode(5)
L = TreeNode(5)

# Level 1 connections (Root to B and G)
A.left = B
A.right = G

# Level 2 connections (Children of B and G)
B.left = C
B.right = F
G.right = H   # 10 appears to be the right child of 15

# Level 3 connections (Leaf nodes)
C.left = D    # 50
C.right = E   # 3

F.left = I    # 1
F.right = J   # 2

H.left = K    # 5
H.right = L   # 5
```

```python
#Pre Order Dfs

arr= []

def pre_order(node):
    if not node:
        return
    
    arr.append(node)
    pre_order(node.left)
    pre_order(node.right)

pre_order(A)
print(arr)
```

```python
def post_order(node) -> int:
    if not node:
        return 0

    left = post_order(node.left)
    right = post_order(node.right)

    return 1 + max(left , right)
    

print(post_order(A))
```

```python
#Question1 : The Problem: Swap every left child with its right child.
arr=[]

def swap_post_order(node):
    if not node:
        return None
    
    left = swap_post_order(node.left)
    right = swap_post_order(node.right)

    node.left = right
    node.right = left

    print(f'right to left: {node.left} , left to right: {node.right} for node {node}')

    arr.append(node)

    return node


swap_post_order(A)
print(arr)
```

```python
#Question2 : The Problem: Given two trees p and q, check if they are identical.

def is_same_tree(p , q):
    if not p and not q:
        return True
    
    if not p or not q:
        return False
    
    if p != q:
        return False

    return is_same_tree(p.left , q.left) and is_same_tree(p.right , q.right)

```

```python
#Question3: Given the root of a binary tree, check whether it is a 
# mirror of itself (i.e., symmetric around its center).

def mirror_tree(l , r):
    if not l and not r:
        return True
    
    if not l or not r:
        return False
    
    if l != r:
        return False

    return mirror_tree(l.left , r.right) and mirror_tree(l.right , r.left)
```

```python
#Question4: Imagine you have two trees printed on transparent plastic sheets. 
# You stack them on top of each other.
# Overlap: If a node exists in both trees at the same spot, 
# add their values (e.g., 2 + 3 = 5).
# No Overlap: If a node exists in only one tree, just keep 
# that node (e.g., None + 4 = 4).
A = TreeNode(1)
B = TreeNode(3)
C = TreeNode(2)

A.left = B
A.right = C

D = TreeNode(2)
E = TreeNode(1)
F = TreeNode(3)
G = TreeNode(4)
H = TreeNode(7)

D.left = E
D.right = F
E.right = G
F.right = H

def merge_tree(p , q):
    if not p and not q:
        return 0
    
    if not p:
        return q
    elif not q:
        return p
    
    p.left = merge_tree(p.left , q.left)
    p.right = merge_tree(p.right , q.right)

    return p + q

merge_tree(A , D)
```
