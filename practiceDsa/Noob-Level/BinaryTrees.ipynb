```python
class TreeNode:
    def __init__(self , node , left=None , right=None) -> None:
        self.node = node
        self.left = left
        self.right = right

    def __repr__(self) -> str:
        return str(self.node)
    

#         2
#      12   15
#     3   2    10
#   50 3 1 2  5  5

A = TreeNode(2)
B = TreeNode(12)
C = TreeNode(3)
D = TreeNode(50)
E = TreeNode(3)
F = TreeNode(2)
G = TreeNode(15)
H = TreeNode(10)
I = TreeNode(1)
J = TreeNode(2)
K = TreeNode(5)
L = TreeNode(5)

# Level 1 connections (Root to B and G)
A.left = B
A.right = G

# Level 2 connections (Children of B and G)
B.left = C
B.right = F
G.right = H   # 10 appears to be the right child of 15

# Level 3 connections (Leaf nodes)
C.left = D    # 50
C.right = E   # 3

F.left = I    # 1
F.right = J   # 2

H.left = K    # 5
H.right = L   # 5
```

```python
#Pre Order Dfs

arr= []

def pre_order(node):
    if not node:
        return
    
    arr.append(node)
    pre_order(node.left)
    pre_order(node.right)

pre_order(A)
print(arr)
```
