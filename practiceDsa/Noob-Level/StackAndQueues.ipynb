```python
#Question5 - The Problem: Given a string like "{[]}", return true if the brackets
# close in the correct order. "{[}]" is false.
from typing import List 
from collections import deque

def ValidParen(stk: List[str]):
    length = len(stk)
    check: str = ')}]'
    arr= deque([])
    pairs = {')': '(', ']': '[', '}': '{'}

    for i in range(0 , length):
        if  stk[i] in '({[':
            arr.append(stk[i])
        elif stk[i] in pairs.keys():
            if not stk:
                return False
            
            topElement = arr.pop()

            if topElement != pairs[stk[i]]:
                return False
            
    return True


stk = ["{" , "[" , "[" , "}"]
print(ValidParen(stk))
```

```python
#Quesion6 - The Problem: Design a stack that supports push, pop, top, and getMin... 
# all in $O(1)$ constant time.

class MinStack:
    def __init__(self , stack) -> None:
        self.min_stack: List[int] = []
        self.stack: List[int] = stack

    def push(self,val):
        self.stack.append(val)

        if self.min_stack[-1] > val:
            self.min_stack.append(val)

    def pop(self):
        top_pop = self.stack.pop()
        
        if top_pop == self.min_stack[-1]:
            self.min_stack.pop()

    def top(self) -> int:
        return self.stack[-1] if self.stack else -1
    
    def getmin(self) -> int:
        return self.min_stack[-1]
```

```python
class Queue:
    def __init__(self) -> None:
        self.stack1:List[int] = []
        self.stack2:List[int] = []

    def push(self,val):
        self.stack1.append(val)
    
    def pop(self) -> int:
        if not self.stack2:
            while self.stack1:
                self.stack2.append(self.stack1.pop())

        return self.stack2.pop()
```

```python
from typing import List

class TempMeasure:
    def __init__(self , arr) -> None:
        self.arr: List[int] = arr

    def days_Till(self) -> List[int]:
        temp: List[int] = []
        result: List[int] = [0] * len(self.arr)

        for i , n in enumerate(self.arr):
            while temp and n > self.arr[temp[-1]]:
                prev_index = temp.pop()

                result[prev_index] = i - prev_index

            temp.append(i)

        return result
```
