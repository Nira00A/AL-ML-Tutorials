```python
#Question1 - The Problem: Given an array of integers nums and an integer target, return 
# indices of the two numbers such that they add up to target. Leetcode 1
from typing import List
import numpy as np

nums = [1,2,3,4,5,6,7]
target = 10

def twoSums(arr: List[int] , target: int) -> List[int] :
    prev_dict : dict = {}

    for i , n in enumerate(arr):
        diff = target - n

        if diff in prev_dict:
            return [prev_dict[diff] , i]
        
        prev_dict[n] = i

    return []

print(twoSums(nums , target))
```

```python
# Double Linked List

class DoubleLinked:
    def __init__(self , val , next=None , prev=None):
        self.next = next
        self.prev = prev
        self.val = val

    def __str__(self):
        return (str(self.val))

def traverse(Head):
    curr = Head
    elements = []
    while curr:
        elements.append(str(curr))
        curr = curr.next
    return " <-> ".join(elements)

Head = DoubleLinked(2)

def insertStart(Head , val):
    new_val = DoubleLinked(val)
    if Head:
        new_val.next = Head
        Head.prev = new_val

    return new_val
    
head = DoubleLinked(2)

head = insertStart(head , 12)
head = insertStart(head , 13)
head = insertStart(head , 14)

print(traverse(head))
```

```python
#Question2 - The Problem: You have an array of prices. Find the max profit 
# (buy low, sell high).You must buy before you sell.
prices = [7,6,4,3,1]


def maxProfit(prices: List[int]):
    min_price = float('inf')
    diff = 0

    for n in prices:
        if min_price > n:
            min_price = n

        elif (n - min_price) > diff:
            diff = n - min_price

    return diff

print(maxProfit(prices))
```

```python
#Question3 - The Problem: Given a string, determine if it is a palindrome 
# (reads the same forward and backward), considering only alphanumeric 
# characters and ignoring cases. (LeetCode #125)

def palindrome(string: str):
    length = len(string)
    l = 0
    r = length - 1
    
    while l < r:
        if string[l] != string[r]:
            return False
        
        l+= 1
        r-= 1

    return True

string = 'aabAa'
print(palindrome(string))

```

```python
#Question4 - The Problem: Find the contiguous subarray 
# (containing at least one number)
#  which has the largest sum.

def kadaneAlgo(arr: List[int]):
    max_so_far = arr[0]
    current_sum = arr[0]

    for i in range(1 , len(arr)):
        current_sum = max(arr[i] , current_sum + arr[i])

        max_so_far = max(max_so_far, current_sum)
        
    return max_so_far
```

```python
#Recurtion practice Merge Sort

def mergeSort(arr: List[int]):
    if len(arr) <= 1:
        return arr
    
    # Divide
    mid = len(arr) / 2
    leftHalf = arr[:mid]
    rightHalf = arr[mid:]

    # Conquer
    sortedLeft = mergeSort(leftHalf)
    sortedRight = mergeSort(rightHalf)

    # Combine
    return merge(sortedLeft , sortedRight)

def merge(left: List[int] , right: List[int]):
    i:int = 0
    j: int = 0
    result: List[int] = []

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:0])
    result.extend(right[j:0])
    return result
```

```python
#Problem4 - The Goal: Calculate $x^n$ (e.g., $2^{10}$).
# The Constraint: Do it faster than $O(N)$. You want $O(\log N)$. (LeetCode #50)

def exN(x:int , n:int):
    if(n == 0):
        return 1
    
    half = exN(x,n //2)
    
    if n % 2 == 0:
        return half * half
    else:
        return x * half * half
    

print(exN(2 , 2))
```
