```python
#Question1 - The Problem: Given an array of integers nums and an integer target, return 
# indices of the two numbers such that they add up to target. Leetcode 1
from typing import List
import numpy as np

nums = [1,2,3,4,5,6,7]
target = 10

def twoSums(arr: List[int] , target: int) -> List[int] :
    prev_dict : dict = {}

    for i , n in enumerate(arr):
        diff = target - n

        if diff in prev_dict:
            return [prev_dict[diff] , i]
        
        prev_dict[n] = i

    return []

print(twoSums(nums , target))
```

```python
# Double Linked List

class DoubleLinked:
    def __init__(self , val , next=None , prev=None):
        self.next = next
        self.prev = prev
        self.val = val

    def __str__(self):
        return (str(self.val))

def traverse(Head):
    curr = Head
    elements = []
    while curr:
        elements.append(str(curr))
        curr = curr.next
    return " <-> ".join(elements)

Head = DoubleLinked(2)

def insertStart(Head , val):
    new_val = DoubleLinked(val)
    if Head:
        new_val.next = Head
        Head.prev = new_val

    return new_val
    
head = DoubleLinked(2)

head = insertStart(head , 12)
head = insertStart(head , 13)
head = insertStart(head , 14)

print(traverse(head))
```

```python
#Question2 - The Problem: You have an array of prices. Find the max profit 
# (buy low, sell high).You must buy before you sell.
prices = [7,6,4,3,1]


def maxProfit(prices: List[int]):
    min_price = float('inf')
    diff = 0

    for n in prices:
        if min_price > n:
            min_price = n

        elif (n - min_price) > diff:
            diff = n - min_price

    return diff

print(maxProfit(prices))
```
